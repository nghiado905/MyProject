```python
import pandas_ta
import pandas as pd
import talib
```

```python
df = pd.read_excel('./FPT.xlsx')
df
```

```python

df = df.drop(['Symbol','Time','TradingDate','Market'],axis=1)
```

```python
df
```

```python
# Define technical strategies:
custom_a = pandas_ta.Strategy(name="First Strategy", ta=[
    {"kind": "sma", "length": 5 },
    {"kind": "sma", "length": 12 },
    {"kind": "rsi","length": 14},
    {"kind": "ema","length": 5},
    {"kind": "ema","length": 12}])


# Add strategies columns to data:
df.ta.strategy(custom_a)
```

```python
df
```

STRATEGY 1

```python
# Then we need to add signal column:
df['entries'] = (df['SMA_5'].shift(1) < df['SMA_12'].shift(1)) & (df['EMA_5'] > df['EMA_12'])
df['entries'] = df['entries'].shift(1).fillna(False)

df
```

BACKTEST FOR STRATEGY 1

```python
import vectorbt as vbt 
```

```python
# Define stop loss and take profit level:
stop_point = 0.1

# Calculate results of the strategy:
bt_pf = vbt.Portfolio.from_signals(df['Close'], 
                                   #entries=df['entries'],
                                   entries=df['entries'],
                                   direction='longonly', 
                                   accumulate='addonly',
                                   sl_stop= stop_point, 
                                   sl_trail=True, 
                                   tp_stop = stop_point, 
                                   fees=0.0015, 
                                   freq= '1D')
```

```python
# Print some statistics of the strategy:
bt_pf.stats() 
```

```python
# Showing signal and results charts:
a = bt_pf.plot()
a.show()
```
