```python
import requests
import time
import csv
from datetime import datetime

# API key ca bn
api_key = "0706188be46c52e0136e9007e8e727d6"
city_name = "Hanoi"
country_code = "VN"  # M quc gia (VN cho Vit Nam)
base_url = "http://api.openweathermap.org/data/2.5/weather?"

# To URL  thc hin yu cu
complete_url = f"{base_url}q={city_name},{country_code}&appid={api_key}"

# Tn tp CSV
file_name = "weather_data1.csv"

# Hm  ghi tiu  (header) vo tp CSV nu cha tn ti
def write_header():
    with open(file_name, 'a', newline='') as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(["Temperature", "Pressure", "Humidity", "Description", "Wind Speed", "Timestamp"])

# Ghi tiu  nu tp CSV cha tn ti hoc trng
try:
    with open(file_name, 'r') as csv_file:
        pass
except FileNotFoundError:
    write_header()

def get_weather_data():
    response = requests.get(complete_url)
    
    if response.status_code == 200:
        weather_data = response.json()
        main = weather_data["main"]
        wind = weather_data["wind"]
        weather_description = weather_data["weather"][0]["description"]

        weather_info = {
            "Temperature": main['temp'] - 273.15,
            "Pressure": main['pressure'],
            "Humidity": main['humidity'],
            "Description": weather_description,
            "Wind Speed": wind['speed'],
            "Timestamp": datetime.now().isoformat()
        }
        
        print(weather_info)
        
        # Lu d liu vo tp CSV
        with open(file_name, 'a', newline='') as csv_file:
            writer = csv.writer(csv_file)
            writer.writerow(weather_info.values())
    else:
        print(f"City not found or there was an error: {response.status_code}")

# Streaming d liu lin tc vi khong thi gian gia cc yu cu
def stream_weather_data(interval=60):
    try:
        while True:
            get_weather_data()
            time.sleep(interval)  # Ch i khong thi gian trc khi gi yu cu tip theo
    except KeyboardInterrupt:
        print("Stopped streaming data")

# Gi hm stream_weather_data vi khong thi gian 60 giy gia cc yu cu
stream_weather_data(60)
```
