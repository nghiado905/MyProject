```python
import requests
import time
import json
from datetime import datetime

# API key của bạn
api_key = "0706188be46c52e0136e9007e8e727d6"
city_name = "Hanoi"
country_code = "VN"  # Mã quốc gia (VN cho Việt Nam)
base_url = "http://api.openweathermap.org/data/2.5/weather?"

# Tạo URL để thực hiện yêu cầu
complete_url = f"{base_url}q={city_name},{country_code}&appid={api_key}"

def get_weather_data():
    response = requests.get(complete_url)
    
    if response.status_code == 200:
        weather_data = response.json()
        main = weather_data["main"]
        wind = weather_data["wind"]
        weather_description = weather_data["weather"][0]["description"]

        weather_info = {
            "Temperature": main['temp'] - 273.15,
            "Pressure": main['pressure'],
            "Humidity": main['humidity'],
            "Description": weather_description,
            "Wind Speed": wind['speed'],
            "Timestamp": datetime.now().isoformat()
        }
        
        print(json.dumps(weather_info, indent=4))
        
        # Lưu dữ liệu vào tệp JSON
        file_name = f"weather_data_{datetime.now().strftime('%Y%m%d%H%M%S')}.json"
        with open(file_name, 'w') as json_file:
            json.dump(weather_info, json_file, indent=4)
    else:
        print(f"City not found or there was an error: {response.status_code}")

# Streaming dữ liệu liên tục với khoảng thời gian giữa các yêu cầu
def stream_weather_data(interval=60):
    try:
        while True:
            get_weather_data()
            time.sleep(interval)  # Chờ đợi khoảng thời gian trước khi gửi yêu cầu tiếp theo
    except KeyboardInterrupt:
        print("Stopped streaming data")

# Gọi hàm stream_weather_data với khoảng thời gian 60 giây giữa các yêu cầu
get_weather_data()
```

```python
import requests
import time
import json
from datetime import datetime, timedelta

# API key của bạn
api_key = "0706188be46c52e0136e9007e8e727d6"

# Tọa độ của thành phố Hà Nội
latitude = 21.0285
longitude = 105.8542

# Thời gian bắt đầu (ngày đầu tiên của năm 2022)
start_date = datetime(2022, 1, 1)

# Số ngày cần lấy dữ liệu
num_days = 1000

# Tạo danh sách các thời gian Unix (Epoch) cho mỗi ngày
timestamps = [int((start_date + timedelta(days=i)).timestamp()) for i in range(num_days)]

# File để lưu dữ liệu
output_file = "weather_data_2022_2023.txt"

# Mở file để ghi
with open(output_file, 'w') as file:
    for i, past_time in enumerate(timestamps):
        # Tạo URL để thực hiện yêu cầu
        base_url = "http://api.openweathermap.org/data/3.0/onecall/timemachine"
        complete_url = f"{base_url}?lat={latitude}&lon={longitude}&dt={past_time}&appid={api_key}"
        
        # Gửi yêu cầu và nhận phản hồi
        response = requests.get(complete_url)
        
        if response.status_code == 200:
            weather_data = response.json()
            
            # Ghi dữ liệu vào file
            file.write(json.dumps(weather_data) + '\n')
            
            print(f"Fetched data for {datetime.fromtimestamp(past_time).strftime('%Y-%m-%d')} (Day {i + 1} of {num_days})")
        else:
            print(f"Error {response.status_code} for {datetime.fromtimestamp(past_time).strftime('%Y-%m-%d')}")
        
        # Chờ một chút để không vượt quá giới hạn tốc độ API
        time.sleep(1)  # Điều chỉnh thời gian nghỉ dựa trên giới hạn tốc độ API của bạn
```
